````python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
````

````python
df=pd.read_csv('SMSSpamCollection.txt',sep='\t',names=['label','messages'])
````

````python
df.shape
````

````python
df.head()
````

### Data Cleaning and Preprocessing

````python
import re
import nltk
nltk.download('stopwords')
````

````python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
````

````python
stop_words=set(stopwords.words('english'))
````

````python
lemma=WordNetLemmatizer()
````

````python
corpus=[]
for i in range(len(df)):
    review=re.sub('[^a-zA-z]',' ',df['messages'][i]).lower().split()
    corpus.append(' '.join([lemma.lemmatize(word) for word in review if word not in stop_words]))
````

````python
corpus
````

````python
y=df['label'].replace({'spam':1,'ham':0})
y
````

````python
y.unique()
````

````python
len(corpus),len(y)
````

### Train Test Split

````python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(corpus,y,test_size=.20,random_state=42)
````

````python
len(xtrain),len(xtest)
````

# Bag Of Words

````python
from sklearn.feature_extraction.text import CountVectorizer
````

````python
cv=CountVectorizer(max_features=2500,ngram_range=(1,2))
````

````python
xtrain
````

````python
x_train=cv.fit_transform(xtrain).toarray()
x_test=cv.transform(xtest).toarray()
````

To prevent data leakage
* xtrain -----> fit_transform
* xtest ------> transfrom 

````python
x_train
````

````python
cv.vocabulary_
````

# Model Building

````python
from sklearn.naive_bayes import MultinomialNB
````

````python
bow_model=MultinomialNB().fit(x_train,ytrain)
````

````python
y_pred=bow_model.predict(x_test)
y_pred
````

````python
from sklearn.metrics import accuracy_score,classification_report
````

````python
accuracy_score(ytest,y_pred)
````

````python
print(classification_report(ytest,y_pred))
````

# Word2Vec

````python
import gensim
from gensim.models import Word2Vec,KeyedVectors
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess
````

````python
words = [simple_preprocess(sent) for sent in corpus]
````

````python
words
````

# Training Word2Vec from scratch

````python
model = Word2Vec(vector_size=100, window=5, min_count=5)
````

````python
model.build_vocab(words)
````

````python
model.train(words, total_examples=model.corpus_count, epochs=model.epochs)
````

````python
# To get all the vocabulary
model.wv.index_to_key
````

````python
# vocabulary size
model.corpus_count
````

````python
model.wv.similar_by_word('good')
````

````python
def avg_word2vec(doc):
    vectors = [model.wv[word] for word in doc if word in model.wv.index_to_key]
    if not vectors:
        return np.zeros(model.vector_size)
    return np.mean(vectors, axis=0)
````

````python
X=[]
for i in range(len(words)):
    X.append(avg_word2vec(words[i]))
````

````python
X
````

````python
len(X)
````

````python
X=pd.DataFrame(X)
````

````python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=.20,random_state=42)
````

````python
from sklearn.ensemble import RandomForestClassifier
````

````python
rf=RandomForestClassifier()
````

````python
rf.fit(xtrain,ytrain)
````

````python
ypred=rf.predict(xtest)
````

````python
from sklearn.metrics import accuracy_score,classification_report
````

````python
accuracy_score(ytest,ypred)
````

````python
print(classification_report(ytest,ypred))
````
